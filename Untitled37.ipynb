```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
A=pd.read_csv('WPP.csv', sep=',')
```

```python
#посмотрим на исходную таблицу
```

```python
A
```

```python
#CBR - рождаемость на тысячу
#CDR - смертность на тысячу
#TFR - кол.-во детей у среднестатистической женщины
#LEx - продолжительность жизни
```

```python
#Поставим задачу изучить динамику изменений данных показателей
```

```python
#Составим таблицы со средними, максимальными и минимальными значениями по годам
```

```python
Amean=A.groupby('Time').mean()[['CBR','CDR','TFR','LEx']]
```

```python
Amean
```

```python
Amax=A.groupby('Time').max()[['CBR','CDR','TFR','LEx']]
```

```python
Amax
```

```python
Amin=A.groupby('Time').min()[['CBR','CDR','TFR','LEx']]
```

```python
Amin
```

```python
#А теперь выведем эту информацию в графическом виде при помощи библиотеки matplotlib, подключённой ранее
```

```python
Date=list(set(A["Time"]))
Date.sort()
```

```python
plt.title("Изменение смертности на тысячу населения")
plt.grid()
plt.plot(Date,Amean["CDR"],Date,Amin["CDR"],Date,Amax["CDR"])
plt.show()
```

```python
plt.title("Изменение рождаемости на тысячу населения")
plt.grid()
plt.plot(Date,Amean["CBR"],Date,Amin["CBR"],Date,Amax["CBR"])
plt.show()
```

```python
plt.title("Изменение фертильности на женщину")
plt.grid()
plt.plot(Date,Amean["TFR"],Date,Amin["TFR"],Date,Amax["TFR"])
plt.show()
```

```python
plt.title("Изменение продолжительности жизни на тысячу населения")
plt.grid()
plt.plot(Date,Amean["LEx"],Date,Amin["LEx"],Date,Amax["LEx"])
plt.show()
```
